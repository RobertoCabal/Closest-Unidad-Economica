```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.decomposition import SparsePCA
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('casas_sample.csv')
_,m = data.shape
colnames = ['supermercados_numero','supermercados_duracion','minisupers_numero','minisupers_duracion',
                     'vinos_licores_numero','vinos_licores_duracion','cerveza_numero','cerveza_duracion',
                     'corporativos_numero','corporativos_duracion','primarias_privadas_numero',
                     'primarias_privadas_duracion','primarias_publicas_numero','primarias_publicas_duracion',
                     'hospitales_privados_numero','hospitales_privados_duracion','hospitales_publicos_numero',
                     'hospitales_publicos_duracion']
colnames = ['CVEGEO','latitud','longitud'] + colnames
data.columns = colnames
```

```python
X = data.iloc[:,3:m].copy().values
pca = PCA(n_components=1)
X_new = pca.fit_transform(X)
data['Comp_Principal_1'] = X_new[:,0]
```

```python
data.to_csv('merida_sample.csv',index=False)
```
